## Análisis sobre posibles outliers

### Se analizará outliers por las ventas totales.

```python
# Calculo de oultliers en las columnas de ventas totales

# Cálculo del primer y tercer cuartil
# Conjunto de datos separando el 25% más bajo y el 25% más alto de los datos
q1 = df['Global_Sales'].quantile(0.25)
q3 = df['Global_Sales'].quantile(0.75)

# Cálculo del rango intercuartílico (IQR)
iqr = q3 - q1

# Límites para detectar outliers
lower_bound = round(q1 - (1.5 *iqr),2)
upper_bound = round(q3 + (1.5 *iqr),2)

# Filtrado de outliers
# Aunque se calcule un lower_bound negativo, se interpretará que no hay outliers inferiores a 0 en ventas
outliers = df[(df['Global_Sales'] < lower_bound)|(df['Global_Sales'] > upper_bound)]
n_outliers = len(outliers)

print(f'lower_bound:{lower_bound}')
print(f'upper_bound:{upper_bound}')
print(f'Q1:{q1}')
print(f'Q3:{q3}')
print(f'IQR:{iqr}')
print(f'outliers encontrados:{n_outliers}')
```

### En el límite inferior sale negativo, este resultado calculado es correcto debido a que la mayoría de juegos tienen ventas muy bajas(entre 0.0 y 0.5). Esto provoca que el tanto Q1 como Q3 sean valores muy pequeños y el IQR sea estrecho.
### Por lo tanto al aplicar la formula de lower_bound, el resultado se negativo. No significa que haya ventas negativas, si no que no hay outliers inferiores a esa distribución, siendo el valor mínimo real 0. Por lo tanto lo vamos a establecer manualmente en 0.

### Mostrar en un Boxplot las ventas globales

```python
lower_bound = 0.00

plt.figure(figsize=(12,6))

# Visualización de outliers en ventas totales
sns.boxplot(x=df['Global_Sales'], color='lightblue')

plt.axvline(q1,color='green', linestyle='--', label='Q1')
plt.axvline(q3,color='orange', linestyle='--', label='Q3')
plt.axvline(lower_bound,color='red', linestyle='--', label='Lower Bound')
plt.axvline(upper_bound,color='purple', linestyle='--', label='Upper Bound')

plt.title('Boxplot de Ventas Globales con Outliers')
plt.xlabel('Ventas Globales (millones de unidades)')
plt.show()
```

### Lo normal es que la mayoría de juegos vendan menos de 1 millon de copias, aunque hay títulos que venden mas. La distribución se puede ver como hay juegos que presentan valores extremadamente grandes. Aunque estos valores no se deben eliminar ya que representan fenómenos reales del mercado.

### Gráfico de densidad

```python
# Cálculo de densidad de las ventas globales
density = df[df['Global_Sales'] >= 0]
plt.figure(figsize=(10,6))
sns.kdeplot(density['Global_Sales'], color='blue', fill=True)
plt.title('Densidad de Ventas Globales')
plt.xlabel('Ventas Globales (millones de unidades)')
plt.ylabel('Densidad')
plt.show()
```

### Se analizará outliers por las ventas en cada región.

```python
# Cálculo de ouliers en las columnas de ventas de cada región

regions = ['NA_Sales','EU_Sales','JP_Sales','Other_Sales']

for region in regions:
    
    df_region = df[region]

    if df_region.empty:
        print(f'Sin datos en {region}')
        continue
    
    q1 = df_region.quantile(0.25)
    q3 = df_region.quantile(0.75)
    iqr = q3-q1

    # Límites para detectar outliers en cada región
    lower_bound = round(q1 - 1.5 * iqr,2)
    upper_bound = round(q3 + 1.5 * iqr,2)

    # Filtrar outliers en cada región

    outliers = df_region[(df_region < lower_bound) | (df_region > upper_bound)]
    n_outliers = len(outliers)

    #Establecer el límite inferior en 0

    lower_bound = 0

    print (f'Región: {region}')
    print('------------------')
    print(f'lower_bound:{lower_bound}')
    print(f'upper_bound:{upper_bound}')
    print(f'Q1:{q1}')
    print(f'Q3:{q3}')
    print(f'IQR:{iqr}')
    print(f'outliers encontrados:{n_outliers}')
    print()
```

### Mostrar en un Bloxpot las ventas de cada región

```python
for region in regions:

    df_region = df[region]
    
    if df_region.empty:
        print(f'Sin datos en {region}')
        continue

    plt.figure(figsize=(12,6))

    # Visualización de outliers en ventas totales
    sns.boxplot(x=df_region, color='lightblue')

    plt.axvline(q1,color='green', linestyle='--', label='Q1')
    plt.axvline(q3,color='orange', linestyle='--', label='Q3')
    plt.axvline(lower_bound,color='red', linestyle='--', label='Lower Bound')
    plt.axvline(upper_bound,color='purple', linestyle='--', label='Upper Bound')

    plt.title(f'Boxplot de Ventas en {region} con Outliers')
    plt.xlabel(f'Ventas en {region}  (millones de unidades)')
    plt.show()
```
